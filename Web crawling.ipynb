### Collect content from links

```python
import requests
from bs4 import BeautifulSoup
```

```python
# Loading web page

url = "https://bp.eosgo.io/"

result = requests.get(url=url)
bs_obj = BeautifulSoup(result.content, "html.parser")

print(bs_obj)
```

```python
# Extract Contents of <div class="If-item">

lf_items = bs_obj.findAll("div", {"class":"lf-item"})

print(lf_items)
```

```python
# Extract link address in <div class="If-item">

hrefs = [div.find("a")['href'] for div in lf_items ]

print(len(hrefs)) #number of links

print(hrefs[0:5])		#partial link value
print(len(hrefs[0:5]))
```

```python
# Extracting the contents of <div class="profile-name">

profile_name = bs_obj.find("div", {'class':'lf-item-info-2'})
print(profile_name)

h4_bp_name = profile_name.find("h4")
print(h4_bp_name)

print(h4_bp_name.text)
```

## Extract Location and Link

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://bp.eosgo.io/listing/eos-csx-2/"

result = requests.get(url)
bs_obj = BeautifulSoup(result.content, "html.parser")

profile_name = bs_obj.find("div", {"class":"profile-name"})

h1_bp_name = profile_name.find("h1")
bp_name = h1_bp_name.text
```

```python
# Extract text

cover_buttons = bs_obj.find("div", {"class":"cover-buttons"})
print(cover_buttons)

button_label = bs_obj.find("span", {"class":"button-label"})
location = button_label.text
print(location)
```

```python
# Button structure check
lis = cover_buttons.findAll("li")
print(lis)

# Check li tag structure
li_tag = lis[1]
print(li_tag)

# print a tag value
a_tag = li_tag.find("a")
print(a_tag)

# print Link Site Address 
link = a_tag['href']
print(link)
```
